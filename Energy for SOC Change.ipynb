#### Quick background

In this script, we will be working with data from SOLON's Tucson, Arizona Tech Park solar installation.

The quantities of interest for our analysis are SOC and BAmps.

SOC is short for "state of charge". It's a percent value that tells you what percentage of max capacity the battery is charged to.

BAmps is short for "battery amps". It is the current flowing from or to the battery at a given time.

There are two questions we want to answer using this data set.

First, we want to know, on average, how many coulombs are required to change the charge of the battery by a single percentage point. This data cannot be easily inferred from the graph simply by visual inspection due to the quantization of the SOC data (all SOC values are integers, which prevents us from accurately computing deltas solely from them) and due to the frequent switchbacks that occur in charging (our battery is cycling between charging and discharging for given intervals).

Second, since the rate at which a battery will charge varies based on its current SOC, we would like to quantify this value for the batteris we are using.

Essentially, the first part of this python notebook is looping through the SOC array and summing the BAmps values between consecutive changes in SOC. The script keeps track of both charging and discharging trends, and denotes multiple SOC changes in the charging and discharging directions using green and blue coloration.

#### Green will denote a trend with only a single count (an SOC change of only a single percentage point before a sign change occurs in the slope of SOC)

#### Blue will denote a trend with multiple integer SOC changes between sign changes in the slope of SOC.

#### Create lists from the dataframes

Dataframes are not meant to be iterrated through, but our analysis requires iteration. We will extract data from several of the columns in the dataframe to be repurposed into lists.

```python
time_list = list(df1.index.values)
soc_list = df1['SOC'].tolist()
bamps_list = df1['BAmps'].tolist()
```

#### Define counter variables that will be used in the loop that will be used to track charging and discharging trends.

So let's just try appending new columns to the data frame with the transformed data that we want.

We have green and blue data sets for both charging & discharging chains.

Create a function 

```python
charging_bamps = []
charging_grn_bamps_delta = []
charging_blue_bamps_delta = []

charging_soc = []
charging_grn_soc = []
charging_blue_soc = []

discharging_bamps = []
discharging_grn_bamps_delta = []
discharging_blue_bamps_delta = []

discharging_soc = []
discharging_grn_soc = []
discharging_blue_soc = []

charging_cons_ctr = 0
discharging_cons_ctr = 0

bamps_cum = 0 #sum of the accumulated amps passing from battery since the last SOC change

#bamps_delta values are created when amps_cum values are passed to them
```

###Create and run loop

```python
for i, k in enumerate(time_list):
    if soc_list[i] > soc_list[i-1]:
        
        charging_bamps.append(bamps_list[i])
        charging_soc.append(soc_list[i])

        charging_cons_ctr = charging_cons_ctr + 1
        discharging_cons_ctr = 0
    
        if charging_cons_ctr == 1:

#             charging_grn_bamps_delta.append(bamps_list[i])
            charging_grn_bamps_delta.append(bamps_cum)
            charging_grn_soc.append(soc_list[i])

        else:

#             charging_blue_bamps.append(bamps_list[i])
            charging_blue_bamps_delta.append(bamps_cum)
            charging_blue_soc.append(soc_list[i])

        bamps_cum = 0
    
    elif soc_list[i] < soc_list[i-1]:
        
        discharging_bamps.append(bamps_list[i])
        discharging_soc.append(soc_list[i])

        discharging_cons_ctr = discharging_cons_ctr + 1
        charging_cons_ctr = 0
    
        if discharging_cons_ctr == 1:

#             discharging_grn_bamps.append(bamps_list[i])
            discharging_grn_bamps_delta.append(bamps_cum)
            discharging_grn_soc.append(soc_list[i])

        else:

#             discharging_blue_bamps.append(bamps_list[i])
            discharging_blue_bamps_delta.append(bamps_cum)
            discharging_blue_soc.append(soc_list[i])

        bamps_cum = 0
    
    else: 
        bamps_cum = bamps_cum + bamps_list[i]
        
        #this condition fires if no SOC change occurs for the given iteration
```

Now we'll perform linear regressions on these two plots

```python
fit_charging = np.polyfit(cbsc,cbbdc,1)
fit_discharging = np.polyfit(dbsc,dbbdc,1)
```

```python
print 'Charging',fit_charging, '\nDischarging', fit_discharging
```

Results from the previous operation returns polynomial coefficients ordered from highest to lowest. (ax+b in this case)

These values quantify the change in charging and discharging rates based on the battery's current SOC value. This answers our second question.
